#### Regression with Sklearn Solver

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
def get_poly_features(data, degree):
    poly_features_transformer = PolynomialFeatures(degree=degree)
    features = poly_features_transformer.fit_transform(data)
    return features
    
    
def OLE_model(degree):
    poly_features_transformer = PolynomialFeatures(degree=degree)
    X_train_poly = get_poly_features(X_train, degree)
    
    regression = LinearRegression()
    model = regression.fit(X_train_poly, y_train)
    return model


def ridge_model(degree, alpha):
    poly_features_transformer = PolynomialFeatures(degree=degree)
    X_train_poly = get_poly_features(X_train, degree)
    
    regression = Ridge(alpha=alpha)
    model = regression.fit(X_train_poly, y_train)
    return model
```

```python
from sklearn.model_selection import cross_validate
best_model = None
best_model_avg_r2 = -1

def evaluate(model, degree):
    X_train_poly = get_poly_features(X_train, degree)
    X_test_poly = get_poly_features(X_test, degree)
    cv_results = cross_validate(model, X_train_poly, y_train, cv=5, return_train_score=True, 
                               scoring=('r2', 'neg_mean_squared_error'))
    print('Train set: ', cv_results['train_r2'])
    print('Test set: ', cv_results['test_r2'])
    avg_r2 = sum(cv_results['test_r2'])/len(cv_results['test_r2'])
    
    global best_model_avg_r2
    global best_model
    if avg_r2 > best_model_avg_r2:
        best_model_avg_r2 = avg_r2 
        best_model = model
```

```python
print('OLE model:')
model_OLE = OLE_model(2)
evaluate(model_OLE, 2)

print('----------------')

print('Ridge model: ')
model_ridge = ridge_model(2, 0.5)
evaluate(model_ridge, 2)
```

```python
model = best_model

X_train_poly = get_poly_features(X_train, 2)
X_test_poly = get_poly_features(X_test, 2)

pred_train = model.predict(X_train_poly)
pred_test = model.predict(X_test_poly)
mse_train = mean_squared_error(y_train, pred_train)
mse_test = mean_squared_error(y_test, pred_test)
print('MSE train: {:.3f}'.format(mse_train))
print('MSE test: {:.3f}'.format(mse_test))
```

#### Regression with SGD

```python
from sklearn.linear_model import SGDRegressor
```

```python
def polynomial_model_SGD(degree): 
    X_train_poly = get_poly_features(X_train[0:-20], degree)
    regression = SGDRegressor(max_iter=100000, loss='squared_loss', validation_fraction=0.1, verbose=0, 
                              eta0=0.0001, learning_rate='constant', tol=1e-3, penalty='l2', alpha=0.02, 
                             random_state=3)
    model = regression.fit(X_train_poly, y_train[0:-20])
    return model
```

```python
def evaluate_SGD(degree):
    model = polynomial_model_SGD(degree)
    X_train_poly = get_poly_features(X_train, degree)
    X_validation_poly = get_poly_features(X_train[-20:-1], degree)
    print('Training set r2: ', model.score(X_train_poly, y_train))
    print('Validation set r2: ', model.score(X_validation_poly, y_train[-20:-1]))
```

#### With TensorFlow

```python
import tensorflow as tf
```

```python
print(X_train[0])
```

```python
degree = 2
X_train_poly = get_poly_features(X_train[0:-20], degree)
y_train_poly = y_train[0:-20].values

X_val_poly = get_poly_features(X_train[-20:-1], degree)
y_val_poly = y_train[-20:-1].values
print(X_train_poly[2].shape)
print(type(y_train_poly))
print(y_train_poly.shape)
```

```python
def compute_result(X, k, n):
    result = tf.add(tf.matmul(X, k), n)
    return result

X = tf.placeholder(tf.float32, shape=(None, X_train_poly[0].shape[0]), name='features')
y = tf.placeholder(tf.float32, shape=(None, 1), name='target')

k = tf.Variable(tf.zeros([91, 1]), name='coeffs', dtype=tf.float32)
n = tf.Variable(tf.zeros(1), name='bias', dtype=tf.float32)


cost = tf.losses.mean_squared_error(y, compute_result(X, k, n))
train_op = tf.train.GradientDescentOptimizer(0.01).minimize(cost)

saver = tf.train.Saver()
bias = None

def train_tf():
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer()) 
        batch_size = 73
        for epoch in range(0, 300):
            print('-------------')
            print('Epoch:', epoch)
            for i in range(0, y_train_poly.shape[0]-batch_size, batch_size):
                sess.run(train_op, feed_dict={X: np.reshape(X_train_poly[i:i+batch_size], (batch_size, 91)), 
                                          y: np.reshape(y_train_poly[i:i+batch_size], (batch_size, 1))})
        
            weigths = sess.run(k)
            bias = sess.run(n)
            x_val = get_poly_features(X_train[-20:-1], degree)
            x_val = tf.constant(x_val, dtype=tf.float32)
            y_val = y_train[-20:-1].values
            MSE = tf.losses.mean_squared_error(np.reshape(y_val, (y_val.shape[0], 1)), compute_result(x_val, weigths, bias))
            error = sess.run(MSE)
            print('MSE:', error)
        saver.save(sess, './model.ckpt')
        
#train_tf()
```

#### On test set

```python
with tf.Session() as sess:
    model_saver = tf.train.import_meta_graph('model.ckpt.meta')
    model_saver.restore(sess, tf.train.latest_checkpoint('./'))
    current_graph = tf.get_default_graph()
    weigths = sess.run(k)
    bias = sess.run(n)
    x_test = get_poly_features(X_test, degree)
    x_test = tf.constant(x_test, dtype=tf.float32)
    y_test = y_test.values
    MSE = tf.losses.mean_squared_error(np.reshape(y_test, (y_test.shape[0], 1)), compute_result(x_test, weigths, bias))
    error = sess.run(MSE)
    print('MSE:', error)    
    #print(current_graph.get_operations())
```
